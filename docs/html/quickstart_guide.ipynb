Retrieve Analyzed Data
----------------------

```python
from nifs.retrieve import AnaData
```

```python
ana = AnaData(80000, 1)
```

```python
ana.diagnostics_list
```

As another options,
diagnostics names are availabe from a [list of analyzed data](http://kaiseki-dev.lhd.nifs.ac.jp/documents/diagnostics/dataname.shtml) or the [search for registerd data](http://egdb.lhd.nifs.ac.jp/dataexplog/registered_data.html) site.
Note that both site is accessible only in NIFS network.

```python
dataset = ana.retrieve("ha1")
dataset
```

```python
dataset["Halph(3O)"].plot()
```
